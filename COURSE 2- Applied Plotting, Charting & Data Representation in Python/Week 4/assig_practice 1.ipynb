```python
ob = np.arange('2016-01-01','2017-01-01',dtype='datetime64[D]')
```

```python
ob = list(map(pd.to_datetime,ob))
```

```python
plt.figure()
xvals = range(len(df1))
plt.bar(xvals,df1['average temperature'],width=1)
plt.bar(xvals,df2_copy['AVEG'],width=1)
```

```python
plt.figure()
plt.bar(df2_copy['AVEG'],'b',alpha=0.34,label='Seattle')
plt.bar(ob,df1['average temperature'],'r',alpha=0.55,label='New York')
plt.legend()
plt.xlabel('In Year 2016')
plt.ylabel('Avg. range of temperature')
plt.title('Temperature comparison between NewYork and Seattle in Year 2016')
          
```
