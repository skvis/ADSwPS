```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df1 = pd.read_csv('weather_data_nyc_centralpark_2016(1).csv')
df2 = pd.read_csv('WeatherSeattle2005-2017.csv')
```

```python
df1 = df1.iloc[91:121]
df1.reset_index(drop=True,inplace=True)
df2 = df2.iloc[4107:4137]
df2.reset_index(drop=True,inplace=True)
```

```python
df2['AVG'] = df2[['TMAX','TMIN']].mean(axis=1)
```

```python
#dates = np.arange('2016-01-01','2017-01-01',dtype='datetime64[D]')
#dates = list(map(pd.to_datetime,dates))
```

```python
plt.figure(figsize=(9,4))
plt.plot(df1['average temperature'],color='b',alpha=0.75,label='New York')
plt.plot(df2['AVG'],color='r',alpha=0.55,label='Seattle')
plt.title('Temperature Comparison Between New York and Seattle in the Months of April-16')
plt.xlabel('Range of Days in a Months ')
plt.ylabel('Average Range of Temperature')
plt.legend()
```
